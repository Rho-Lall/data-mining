## Descriptive Statistics (Measuring Central Tendencies & Dispersion)

### Summarize important elements in a dataset, to understand what it looks like

### Univariate

- ### Measures of Central Tendency
    - #### mean, median, mode
    - #### gemetric & harmonic means (ratios & rates)

- ### Measures of Dispersion
    - #### range (max - min)
    - #### Inter-Quartile Range - middle 50% of data
    - #### Standard Deviation & Variance
    
- ### Confidence Intervals
     - #### Sampling Distribution: Probability distribution of a population statistic (mean), given a particular sample.
     - #### Sample mean is the best, unbiased estimator of the population mean
     - #### Confidence is dependant on sample variance & sampe size
         - mean = sample mean
         - varaince ~ sample variance / n (n=# of observations in the sample)
         - std.dev ~ sample stdv / sqrt(n)

- ### Skewness & Kurtosis
    - #### Skewnesss looks at distortions due to caps and floors in the data
    - #### Kurtosis is a measure of tail risk on a normal distribution, or likelyhood of extream events.
    

 
### Bivariate

- ### Correlation & Covariance
    - Correlation is a normalized version of covariance (-1,1)


### Multivariate

- ### Correlation Matrix & Covariance Matrix


## Inferencial Statistics

### Explain the elements of the data in terms of relationships with other elements

- ### Hypothesis Testing - Answer the question is your story supported by the data?
- ### Model Fitting - predictive analytics

# Import Data

```python
#I am not going to analize CaseOrder, but if there are outliers I want to be able to identify them.
data = pd.read_csv('../data/churn_wgu/clean.csv')
data.head(5)
```

```python
# returns a count of null values
data.isnull().sum()
```

```python
unwanted = data.columns[data.columns.str.startswith('Survey_')]
data.drop(unwanted, axis=1, inplace=True)
model_data = data.drop(['CaseOrder', 'Lat', 'Lng', ], axis=1)
model_data
```

# Summary Statistics: Tenure

```python
data.Tenure.describe().to_frame()
```

```python
data.Tenure.median()
```

```python
# .mode
data.Tenure.value_counts().to_frame()
```

```python
plt.figure(figsize=(10,10))
data.Tenure.hist(bins=72, align='left')
plt.axvline(data.Tenure.mean(),label='mean', color='r')
plt.axvline(data.Tenure.median(),label='median', color='b')
plt.axvline(data.Tenure.quantile(.20),label='20th Percentile', color='yellow')
plt.axvline(data.Tenure.quantile(.80),label='80th Percentile', color='yellow')
plt.title('Histogram: Tenure')
plt.xlabel('Months')
plt.ylabel('Customers')
plt.xlim([0,72])
#plt.ylim([0,432])
plt.xticks([0,6,12,18,24,30,36,42,48,54,60,66,72])
plt.legend();
```

```python
data.Tenure.quantile(.25)
```

```python
#dispersion
display(data.Tenure.var())
display(data.Tenure.std())
```

```python
plt.figure(figsize=(10,10))
data.Tenure.plot(kind='box',
                title='Boxplot: Tenure'
);
```

```python
# I bimodal distribution suggests that there are possibly two distinct groups represented in this data
plt.figure(figsize=(10,10))
data.Tenure.plot(kind='kde',
                title='Kernal Density Estimation: Tenure'
);
```

```python
display(data.Tenure.skew())
display(data.Tenure.kurtosis()) # extreame values are less likely than a normal dist
```

# Summary Statistics: Independent Variables

```python
#Correlation Matrix
corr = model_data.corr()
f, ax = plt.subplots(figsize=(11, 9))
ax.set_title('Independent Variables: Correlation Matrix')
mask = np.triu(np.ones_like(corr, dtype=bool))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.99, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5});
```

```python
pd.set_option('display.max_rows', None)
c = model_data.corr().abs()
s = c.unstack()
so = s.sort_values(kind='quicksort', ascending=False)
so[42:52]
```

# StatsModels

```python
from scipy import stats
from statsmodels.stats.weightstats import DescrStatsW
```

```python
#range
np.ptp(data.Tenure)
```

```python
#Descriptive Statistics
tenure_stats = DescrStatsW(data.Tenure)
tenure_stats.quantile(.25).values[0]
```

```python
# Interquartile Range
stats.iqr(data.Tenure)
```
