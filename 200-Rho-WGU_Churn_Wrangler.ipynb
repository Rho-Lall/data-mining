```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('display.max_columns', None)
```

```python
raw = pd.read_csv('../data/churn_wgu/churn_clean.csv')
```

```python
raw.head()
```

```python
# Drop Unique Identifiers, Customer_id, Interaction and UID as we already have a unique idendentifier, index.
raw = raw.drop(['Customer_id', 'Interaction', 'UID'], axis=1)
```

```python
# Drop address data as we have lat/lng. This is redundant and lat/lng is preferable as address data is too granular to offer value as a categorical feature
raw = raw.drop(['City', 'State', 'County', 'Zip'], axis=1)
raw
```

```python
# review values for Churn to ensure there are no NaN or unexpected values
raw.Churn.value_counts()
```

```python
# encode Churn as a Dummy Variable, Churned_Away
raw['Churned_Away'] = raw.Churn.map({'Yes': 1, 'No': 0})
raw = raw.drop(['Churn'], axis=1)
raw
```

```python
# review values for Gender to look for NaN or unexpected values
raw.Gender.value_counts()
```

```python
# prepare dummy variables for Gender. 
gender_dummies = pd.get_dummies(raw.Gender, prefix='Gender', drop_first=True)
raw = pd.concat([raw, gender_dummies], axis=1)
raw = raw.drop(['Gender'], axis=1)
raw
```

```python
# Area may be useful as a categorical feature
raw.Area.value_counts()
```

```python
# prepare dummy variables for Area. 
area_dummies = pd.get_dummies(raw.Area, prefix='Area')
raw = pd.concat([raw, area_dummies], axis=1)
raw = raw.drop(['Area', 'Area_Rural'], axis=1)
raw
```

```python
# review values for TimeZone to look for NaN or unexpected values
# There are too many for this to be useful on a first pass
# Group these into the four main continental timezones and other
raw.TimeZone.value_counts()
```

```python
raw['Time_zone'] = raw.TimeZone.map({
    'America/New_York': 'Eastern',
    'America/Chicago': 'Central',
    'America/Los_Angeles': 'Pacific',
    'America/Denver': 'Mountain',
    'America/Detroit': 'Eastern',
    'America/Indiana/Indianapolis': 'Eastern',
    'America/Phoenix': 'Mountain',
    'America/Boise': 'Mountain',
    'America/Anchorage': 'Other',
    'America/Puerto_Rico': 'Other',
    'Pacific/Honolulu': 'Other',
    'America/Menominee': 'Central',
    'America/Nome': 'Other',
    'America/Kentucky/Louisville': 'Eastern',
    'America/Sitka': 'Other',
    'America/Indiana/Tell_City': 'Central',
    'America/Indiana/Vincennes': 'Eastern',
    'America/Toronto': 'Eastern',
    'America/Indiana/Petersburg': 'Eastern',
    'America/North_Dakota/New_Salem': 'Central',
    'America/Juneau': 'Other',
    'America/Indiana/Knox': 'Central',
    'America/Indiana/Marengo': 'Eastern',
    'America/Ojinaga': 'Other',
    'America/Indiana/Winamac': 'Eastern'
    
    })
raw = raw.drop(['TimeZone'], axis=1)
raw
```

```python
# review values for Time_zone to look for NaN or unexpected values
raw.Time_zone.value_counts()
```

```python
# prepare dummy variables for Time_zone
tz_dummies = pd.get_dummies(raw.Time_zone, prefix='TimeZone')
raw = pd.concat([raw, tz_dummies], axis=1)
raw = raw.drop(['Time_zone', 'TimeZone_Other'], axis=1)
raw
```
