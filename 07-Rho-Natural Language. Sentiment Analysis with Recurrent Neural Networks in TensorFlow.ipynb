**How Best Can Words be Represented as Numeric Data?**

There are three methods of encoding text: one-hot, frequency based, & prediction based.


**One Hot.** Each document (sentenece) will be a tensor of rank 1 with as many elements as there are words in the entire vocabulary. IE d = [0,0,0,0,1,0]

Flaws in One Hot encoding are:
- (1) it requires a large vocabulary and enormous feature vectors
- (2) represents unordered view of words and looses all context
- (3) representation is binary, we don't know how often a word appears in the document (loss of frequency info)
- (4) does not capture any semantic information or relationship between words

**Frequency Based** There are three types of frequency based encodings: count, TF-IDF, & co-occurence. Count captures how often a word occurs in a document (frequency). 


Flaws of Count encoding:
- (1) it requires a large vocabulary and enormous feature vectors
- (2) represents unordered view of words and looses all context


TF-IDF. Captures how often a word occurs in a document as well as the entire corpus.
